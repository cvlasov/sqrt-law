```python
import matplotlib.pyplot as plt
```

```python
x = [7.68, 83.6652, 162.5088, 240.8448, 314.5728, 398.1312, 479.3088, 554.88, 635.9808, 707.7888]

# O(1)
y1 = [1-0.0,1-0.0457,1-0.1064,1-0.1638,1-0.2023,1-0.2352,1-0.2625,1-0.5,1-0.2907,1-0.5]

# O(sqrt(n))
y2 = [1-0.0674,1-0.1574,1-0.1842,1-0.1918,1-0.2023,1-0.2054,1-0.2115,1-0.2053,1-0.2054,1-0.2081]

# O(sqrt(n) * log(n))
y3 = [1-0.1268,1-0.1809,1-0.1953,1-0.5,1-0.2023,1-0.5,1-0.5,1-0.5,1-0.5,1-0.5]

# O(n)
y4 = [1-0.4205,1-0.3237,1-0.5,1-0.5,1-0.2023,1-0.5,1-0.5,1-0.5,1-0.5,1-0.5]
```

```python
plt.figure(figsize=(12,10))

plt.subplot(2,2,1)
plt.plot(x, y1, 'r+:', label='constant 1')
plt.axis([0,800,0.5,1.0])
plt.xlabel(r'Cover size N in pixels ($\times 10^4$)')
plt.ylabel(r'$1-P_E$')
plt.title(r'$O(1)$')
plt.legend()

plt.subplot(2,2,2)
plt.plot(x, y2, 'r+:', label='constant 1')
plt.axis([0,800,0.5,1.0])
plt.xlabel(r'Cover size N in pixels ($\times 10^4$)')
plt.ylabel(r'$1-P_E$')
plt.title(r'$O(\sqrt{N})$')
plt.legend()

plt.subplot(2,2,3)
plt.plot(x, y3, 'r+:', label='constant 1')
plt.axis([0,800,0.5,1.0])
plt.xlabel(r'Cover size N in pixels ($\times 10^4$)')
plt.ylabel(r'$1-P_E$')
plt.title(r'$O(\sqrt{N}\log{N})$')
plt.legend()

plt.subplot(2,2,4)
plt.plot(x, y4, 'r+:', label='constant 1')
plt.axis([0,800,0.5,1.0])
plt.xlabel(r'Cover size N in pixels ($\times 10^4$)')
plt.ylabel(r'$1-P_E$')
plt.title(r'$O(N)$')
plt.legend()

plt.tight_layout()
plt.savefig('graph_18_02_2019.png')
plt.show()
```
